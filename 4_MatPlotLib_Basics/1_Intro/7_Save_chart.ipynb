```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone","Car","Others"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,
        autopct="%1.1f%%",radius=1.5)
plt.savefig("pie_chart.jpg",transparent=True)
```
