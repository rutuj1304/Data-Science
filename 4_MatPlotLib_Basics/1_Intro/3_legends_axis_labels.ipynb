```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,42,39]
min_t = [30,41,39,40,32,31,29]
avg_t = [40,46,46,44,40,37,34]

plt.plot(days, max_t, label = "Max Temp")
plt.plot(days, min_t, label = "Min Temp")
plt.plot(days, avg_t, label = "Avg Temp")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
# plt.legend(loc="best")
plt.legend(shadow=True,fontsize="large")
```
