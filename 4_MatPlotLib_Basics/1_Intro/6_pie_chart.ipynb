```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone","Car","Others"]

plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,
        autopct="%1.1f%%",radius=1.5)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,
        autopct="%1.1f%%",radius=1.5,explode=[0,0,0,0.1,0.2])
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,
        autopct="%1.1f%%",radius=1.5,explode=[0.1,0.2,0.2,0.1,0.2])
plt.axis("equal")
plt.show()
```
