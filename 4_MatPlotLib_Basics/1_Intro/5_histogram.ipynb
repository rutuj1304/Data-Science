# Histogram of Blood Sugar Levels
x contains variable and y axis will be frequency of each value
By default number of bins is set to 10
rwidth is the width of the bars


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8)
```

```python
plt.hist(blood_sugar,rwidth=0.4,bins=4)
```

```python

plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
         color="g")
```

```python
plt.hist(blood_sugar,rwidth=0.4,bins=4,histtype='stepfilled')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67,98,89,120,133,150,84,79,100,120,112,100]

plt.hist([blood_sugar_men,blood_sugar_women],rwidth=0.75,
         bins=[80,100,125,150],color=['green','orange'],label=['Men','Women'])
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
        histtype="step")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
        histtype="stepfilled")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
        histtype="barstacked",color="g")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
        histtype="bar",color="g")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],
        orientation="horizontal",color="g")
```
