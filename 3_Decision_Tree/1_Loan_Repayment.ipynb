```python
columns = ['Gender','Dependents','LoanAmount','Loan_Amount_Term']
```

```python
data = data.dropna(subset=columns)
```

```python
data.isnull().sum()
```

```python
data['Self_Employed'].mode()[0]
```

```python
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
data['Gender'].unique()
```

```python
data['Self_Employed'].unique()
```

```python
data['Credit_History'].mode()[0]
```

```python
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['Dependents'] = data['Dependents'].replace(to_replace='3+',value='4')
```

```python
data['Dependents'].unique()
```

```python
data['Loan_Status'].unique()
```

```python
data.head()
```

```python
data['Gender'] = data['Gender'].map({'Male':0,'Female':1}).astype('int')
```

```python
data.head()
```

```python
data['Married'] = data['Married'].map({'Yes':0,'No':1}).astype('int')

data['Education'] = data['Education'].map({'Graduate':1,'Not Graduate':0}).astype('int')

data['Self_Employed'] = data['Self_Employed'].map({'Yes':1,'No':0}).astype('int')

```

```python
data.head()
```

```python
data['Property_Area'] = data['Property_Area'].map({'Rural':0,'Urban':1,'Semiurban':2}).astype('int')

data['Loan_Status'] = data['Loan_Status'].map({'Y':1,'N':0}).astype('int')
```

```python
data.head()
```

```python
data = data.drop('Loan_ID',axis=1)
```

```python
data.head()
```

```python
x = data.drop('Loan_Status',axis=1)
```

```python
x
```

```python
y = data['Loan_Status']
```

```python
y
```

feature scaling

```python
data.head()
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']
```

```python
from sklearn.preprocessing import StandardScaler

st = StandardScaler()

x[cols] = st.fit_transform(x[cols])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8) 
```

```python
x_train
```

```python
y_train
```

```python
models = [DecisionTreeClassifier()]
for model in models:
    model.fit(x_train,y_train)
```

```python
model_Names = ['Decision Tree']
for model , name in zip(models,model_Names):
    print(name + ":{:.4f}%".format(model.score(x_test,y_test) * 100))
```
