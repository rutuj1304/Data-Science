# Best and worst movies

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
df_best = df[["poster_path","title","budget_musd","revenue_musd",
              "vote_count","vote_average","popularity"]].copy()
df_best
```

```python
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
df_best.columns = ["Poster_Path","Title","Budget","Revenue","Votes","Average Rating","Popularity","Prfit","ROI"]
```

```python
df_best.set_index("Title",inplace=True)
```

```python
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending=False)
```

```python
df_best.Budget.fillna(0,inplace=True)
df_best.Votes.fillna(0,inplace=True)
```

```python
df_best.info()
```

```python

def best_worst(n,by,ascending=False,min_bud=0,min_vote=0):
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_vote),["Poster_Path",by]].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

Movies Top 5 - Highest Revenue

```python
best_worst(n=5,by="Revenue")
```

Movies Top 5 - Highest Profit

```python
best_worst(5 , "Budget")
```

```python
act = act.stack().reset_index(level=1,drop=True).to_frame()
act
```

```python
act.columns = ["Actor"]
```

```python
act
```

```python
act = act.merge(df[["title","revenue_musd","vote_average","popularity"]],how="left",left_index=True,right_index=True)
```

```python
act
```

```python
act.Actor.unique()
```

```python
act.Actor.value_counts().head(20)
```

```python
plt.figure(figsize=(12,8))
act.Actor.value_counts().head(20).plot(kind="bar")
plt.title("Most Active Actors",fontsize=20,color='r')
plt.ylabel("Number of Movies",fontsize=15,color='b')
plt.show()
```

```python
agg = act.groupby("Actor").agg(
    Total_Revenue = ("revenue_musd","sum"),
    Mean_Revenue = ("revenue_musd","mean"),
    Mean_Rating = ("vote_average","mean"),
    Mean_pop = ("popularity","mean"),
    Total_Movies = ("Actor","count")
)
```

```python
agg
```

```python
agg.nlargest(10,"Total_Movies")
```

```python
agg.nlargest(10,"Total_Revenue")
```

```python
plt.figure(figsize=(12,8))
agg.Total_Revenue.nlargest(10).plot(kind="line",fontsize=15,color='b')
plt.title("Top 10 Highest Grossing Actors",fontsize=20,color='r')
plt.ylabel("Total Revenue",fontsize=15,color='b')
```
