```python
#medal per country bar chart
medal_per_country = pd.crosstab(index=olympics.Country,columns=olympics.Medal,
                                margins=True,margins_name="Total")
medal_per_country
```

```python
medal_per_country.drop("Total",axis=0,inplace=True)
```

```python
medal_per_country
```

```python
medal_per_country=medal_per_country.sort_values(by=["Total"],ascending=False)
```

```python
#summer and winter edition cross tab
sum_vs_win = pd.crosstab(olympics.Country,olympics.Edition)
```

```python
sum_vs_win
```

```python
#men and women cross tab
men_vs_women = pd.crosstab(olympics.Country,olympics.Gender)
```

```python
men_vs_women
```

```python
medals_by_cats = pd.concat([sum_vs_win,men_vs_women],axis=1)
medals_by_cats
```

```python
medals_by_cats["Total"] = medals_by_cats.Summer + medals_by_cats.Winter
medals_by_cats
```

```python
medals_by_cats.sort_values("Total",ascending=False,inplace=True)
medals_by_cats
```

```python
ranks = medals_by_cats.rank(ascending=False)
ranks
```
