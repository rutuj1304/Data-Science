```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Dataset = pd.read_csv("Titanic_DataSet.csv")
Dataset
```

```python
Dataset.info()
```

```python
#rename the column name
Dataset.columns = ["Name","Survived","Gender","Age","Class"]
Dataset.head()
```

```python
#Find out max age pax and display it
#Find out min age pax and display it
max_age = Dataset.groupby("Age").max().reset_index().sort_values(by="Age", ascending=False)
max_age.head(1)
```

```python
min_age = Dataset.groupby("Age").min().reset_index().sort_values(by="Age", ascending=True)
min_age.head(1)
```

```python
Dataset.loc[Dataset.Age==Dataset.Age.max()]
```

```python
Dataset.loc[Dataset.Age==Dataset.Age.min()]
```

```python
#Display passengers by class wise
Dataset.groupby("Class")['Name'].count().reset_index()
```

```python
#diplay survieved  yes or no
Dataset['Survived'].value_counts().reset_index()
```

```python
#display min age and max age from survived 
Dataset_survived = Dataset.groupby('Survived')
Dataset_survived.groups
```

```python
len(Dataset_survived.get_group('yes'))
```

```python
survivor = Dataset_survived.get_group('yes')
```

```python
survivor.loc[survivor.Age==survivor.Age.min()]
```

```python
survivor.loc[survivor.Age==survivor.Age.max()]
```

```python
#non survived max age and min age passengers
non_survivor = Dataset_survived.get_group('no')
non_survivor.loc[non_survivor.Age==non_survivor.Age.min()]
```

```python
non_survivor.loc[non_survivor.Age==non_survivor.Age.max()]
```
