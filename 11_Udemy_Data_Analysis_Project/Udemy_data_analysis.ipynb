```python

data.groupby('course_title')['num_subscribers'].sum().reset_index().sort_values(by='num_subscribers', ascending=False).head(20)
```

```python
#List of selling courses
data.groupby('course_title')['num_subscribers'].sum().reset_index().sort_values(by='num_subscribers', ascending=True).head(20)
```

```python
#Show all courses of graphic design where price is below or equal to 100
data[(data.subject == 'Graphic Design') & (data.price <= '100')]
```

```python
# Show courses that are published in year 2015
data.info()
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])
data['year']=data['published_timestamp'].dt.year
data[data['year']==2015]
```

```python
#in 2015, find course title and no. of subscibers
new_df = data[data['year']==2015]
new_df.groupby('course_title')['num_subscribers'].sum().reset_index().sort_values(by='num_subscribers', ascending=False).head(10)
```
