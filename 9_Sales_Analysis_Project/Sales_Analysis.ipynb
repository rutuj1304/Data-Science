```python
df['Customer Age'] = df['Customer Age'].astype(int)
df['Year'] = df['Year'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df
```

```python
#Extract year from date
df['Year'] = df['Date'].dt.year
```

```python
df
```

```python
#mon-year value to be stored in new column
df['Year-Month'] = df['Date'].dt.strftime('%Y-%m')
df
```

```python
# Column Creation
df['Profit'] = df['Revenue']-df['Cost']
df.head()
```

```python
df.info()
```

```python
df['Profit_Margins'] = round((df['Profit']/df['Revenue'])*100,2)
```

```python
df.head()
```

```python
#display revenue for particular month
Month_Revenue = df.groupby('Year-Month')['Revenue'].sum().reset_index()
Month_Revenue
```

```python
#year month and cost 
Year_Month_Cost = df.groupby('Year-Month')['Cost'].sum().reset_index()
```

```python
Year_Month_Cost
```

```python
#group by month and display cost,revenue,profit with respective months
Grouped = df.groupby('Year-Month')[['Cost','Revenue','Profit']].sum().reset_index()
Grouped
```

```python
import plotly.express as px
```

```python
fig = px.line(Grouped, x="Year-Month", y=['Cost','Revenue','Profit'], title='Monthly Performence')
fig.update_xaxes(title="Year-Month")
fig.update_yaxes(title="Amount($)")
fig.show()
```

```python
# group by subcategory and calculate total quantity sold
#and Display order by quantity
Subcategory_Sales = df.groupby('Sub Category')['Quantity'].sum().reset_index().sort_values('Quantity', ascending=False)
Subcategory_Sales
```

```python
# plot the bar chart using plotly
fig = px.bar(Subcategory_Sales, x="Quantity", y="Sub Category", title='	Quantity BY Sub Category',text_auto='.2s')
fig.show()
```

```python
# group by subcategory and calculate total Profit 

Subcategory_profit = df.groupby('Sub Category')['Profit'].sum().reset_index()
Subcategory_profit
```

```python
# plot the bar chart using plotly
fig = px.bar(Subcategory_profit, x="Profit", y="Sub Category", title='Profit BY Sub Category',text_auto='.2s')
fig.update_xaxes(title="Profit($)")
fig.show()
```

```python
# group by subcategory and calculate total Profit_Margins

Subcategory_Profit_margins = df.groupby('Sub Category')['Profit_Margins'].mean().round(2).reset_index()
Subcategory_Profit_margins
```

```python
# plot the bar chart using plotly
fig = px.bar(Subcategory_Profit_margins, x="Profit_Margins", y="Sub Category", title='Profit_Margins BY Sub Category',text_auto='.2s')
fig.show()
```

```python
#Plot histogram on Customer Age in dataframe
fig = px.histogram(df, x="Customer Age", title='Customer Age Histogram')
fig.show()
```

```python
#Group by Customer Age and Product Category
#Display Total sold quantity
df_grouped = df.groupby(['Customer Age','Product Category'])['Quantity'].sum().reset_index()
```

```python
df_grouped
```

```python
fig = px.bar(df_grouped, x="Customer Age", y="Quantity", title='Quantity BY Age',color='Product Category',text_auto='.2s')
fig.show()
```
