```python
# male and females in customers
df.Gender.value_counts().to_frame()
```

```python
ax = sns.countplot(x="Gender",data=df)
for labels in ax.containers:
    ax.bar_label(labels)
```

```python
#Gender wise sales
gen_sales = df.groupby('Gender')['Amount'].sum().reset_index().sort_values(by="Amount",ascending=False)
```

```python
gen_sales
```

```python
#create bar chart
ax1 = sns.barplot(x="Gender",y="Amount",data=gen_sales)
for labels in ax1.containers:
    ax1.bar_label(labels)
```

```python
#age-group = 26-35 and sort by gender
df2 = df[['Age Group','Gender']].value_counts().to_frame().sort_values(by="Gender",ascending=False)
df2
```

```python
ax = sns.countplot(x="Age Group",hue="Gender",data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
#total sales amount using age group
age_sales = df.groupby('Age Group')['Amount'].sum().reset_index().sort_values(by="Amount",ascending=False)
age_sales
```

```python
ax = sns.barplot(x="Age Group",y="Amount",data=age_sales)
#instead using for loop , display labels for diagram
ax.bar_label(ax.containers[0],labels=age_sales['Amount'])
```

```python
#Total number of orders of top 10 states
sales_state = df.groupby('State')['Orders'].sum().reset_index().sort_values(by="Orders",ascending=False).head(10)

sales_state
sns.set(rc={'figure.figsize':(15,8)})
ax = sns.barplot(x="State",y="Orders",data=sales_state)
ax.bar_label(ax.containers[0],labels=sales_state['Orders'])
```

```python
#Total sales amount of top 10 states
sales_amount = df.groupby('State')['Amount'].sum().reset_index().sort_values(by="Amount",ascending=False).head(10)

sales_amount
sns.set(rc={'figure.figsize':(15,8)})
ax = sns.barplot(x="State",y="Amount",data=sales_amount)
ax.bar_label(ax.containers[0],labels=sales_amount['Amount'])
```

```python
#display total count of marital status column
ax = sns.countplot(x="Marital_Status",data=df)
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(15,8)})
```

```python
#Total Sales > Gender , Marital Status
#Display male and female in marital status
#display total sales according gender
marital_status = df.groupby(['Marital_Status','Gender'])['Amount'].count().reset_index().sort_values(by="Amount",ascending=False)

marital_status
sns.set(rc={'figure.figsize':(15,8)})
ax = sns.barplot(x="Marital_Status",y="Amount",data=marital_status,hue="Gender")
for i in ax.containers: ax.bar_label(i)
```

```python
# Display Total Sales by Occupation
occupation = df.groupby(['Occupation'])['Amount'].sum().reset_index().sort_values(by="Amount",ascending=False)
occupation
sns.set(rc={'figure.figsize':(12,8)})
ax = sns.barplot(x="Occupation",y="Amount",data=occupation)
ax.bar_label(ax.containers[0],labels=occupation['Amount'])
```

```python
#Total Orders places by each Occupaton Sector
occupation_orders = df.groupby(['Occupation'])['Amount'].count().reset_index().sort_values(by="Amount",ascending=False)
occupation_orders
sns.set(rc={'figure.figsize':(12,8)})
ax = sns.barplot(x="Occupation",y="Amount",data=occupation_orders)
ax.bar_label(ax.containers[0],labels=occupation_orders['Amount'])
```

```python
#Group By Product Category and Amount
product_category = df.groupby(['Product_Category'])['Amount'].sum().reset_index().sort_values(by="Amount",ascending=False).head(10)
product_category
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x="Product_Category",y="Amount",data=product_category)
ax.bar_label(ax.containers[0],labels=product_category['Amount'])
```
